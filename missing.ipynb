```python
import pandas as pd
```

```python
missing = pd.read_pickle('data/missing-trials.pkl')
missing
```

```python
# Parameter settings that failed on every dataset
for methodName, methodGroup in missing.groupby('method'):
    for pString, paramGroup in methodGroup.groupby('parameters'):
        u = len(paramGroup['dataset'].unique()) # get all datasets this parameter setting failed on
        if u == 166:
            print(methodName, u, pString)
```

```python
for pString, group in missing[missing['method'] == 'SVC'].groupby('parameters'):
    u = len(group['dataset'].unique())
    print('SVC', u, pString)
```

```python
missing = pd.read_pickle('data/missing-trials-7-30.pkl')
missing
```

```python
print("Timed out: ", len(missing[missing['reason'] == 'Timed out!']))
print("Errored: ", len(missing[missing['reason'] != 'Timed out!']))
```

```python
for classifier, group in missing[missing['reason'] != 'Timed out!'].groupby('classifier'):
    print(classifier, len(group['datasets'].unique()))
```

```python
for dataset, group in missing[missing['reason'] == 'Timed out!'].groupby('datasets'):
    print(dataset, list(group['classifier'].unique()))
```
